```python
row_list=[]
for idx, r in ratings.iterrows():
    row = r.values[0].split(" ")
    try:
        row_list.append((row[6],row[8],row[11], ' '.join(row[12:-1])[1:], row[-1][1:-1]))
    except:
        pass
len(row_list)
```

```python
ratings = pd.DataFrame(row_list, columns=['distribution', 'votes', 'rank', 'title', 'year'])
ratings=ratings.drop_duplicates()
ratings.to_csv("ratings.csv", index=False)
ratings.head()
```

```python
top250=ratings.head(250)
```

```python
vals =top250.values
vals[0]
```

```python
N=250
adj=np.zeros((N, N))
vals =top250.values
for i in range(N):
    for j in range(i+1, N):
        if abs(top250.ix[i][-1]-top250.ix[j][-1])<1:
            adj[i][j]=1
            adj[j][i]=1
adj
```

```python
movies = []
for i in range(250):
    title = top250.ix[i]['title']
    related = []
    for j in range(i+1, N):
        if adj[i][j]==1:
            related.append(top250.ix[j]['title'])
    movies.append({"title": title, 'related': related})
movies
```

```python
nodes=[]
edges=[]
for item in related[0:30]:
    if item['title'][0]=='"':
        continue
    nodes.append({'data': {"id": item['title']}})
    for rel in item['related']:
        if rel[0]=='"':
            continue
        if rel not in nodes:
            nodes.append({'data': {"id": rel}})
        edges.append({'data': {"source": item['title'], "target": rel}})
elements={'nodes': nodes, 'edges': edges}
elements
```

```python

with open('elements.json', 'w') as outfile:
    json.dump(elements, outfile, ensure_ascii=False)
```
