```python
row_list=[]
for idx, r in ratings.iterrows():
    row = r.values[0].split(" ")
    try:
        row_list.append((row[6],row[8],row[11], ' '.join(row[12:-1])[1:], row[-1][1:-1]))
    except:
        pass
len(row_list)
```

```python
ratings = pd.DataFrame(row_list, columns=['distribution', 'votes', 'rank', 'title', 'year'])
ratings=ratings.drop_duplicates()
ratings.to_csv("ratings.csv", index=False)
ratings.head()
```
